## Create Pipeline

````python
# TODO: Create a pipeline for each set.
# Resize, normalize and create batches (to reduce number of images that are processed at the same time)
# Definitions
imgShape = 224 # width/height for the images to be used by the model
batchSize = 64
shuffleSize = numTrainExamples//4 # integer value for size of shuffle buffer (training-samples-only)

# Preparation method (resize, normalize) to be applied to dataset splits by .map()
def prepareImg(image, label):
    image = tf.image.resize(image, (imgShape, imgShape)) # Overwrite image with resized image
    image /= 255.0 # Normalize to maximum value of byte ((2^8)-1 = 255), use float to get scaled values as floats too
    return image, label

# Apply preparation method and create batches
# .cache(): only first reading will generate data; subsequent iterations will read from cache
# .shuffle(): shuffle the elements of the dataset randomly from a buffer
# .map(): maps passed function across all elements of the dataset 
# .prefetch(): creates dataset that prefetches elements from dataset (buffer_size = 1)
trainingBatches = training_set.cache().shuffle(shuffleSize).batch(batchSize).map(prepareImg).prefetch(1)
validationBatches = training_set.cache().batch(batchSize).map(prepareImg).prefetch(1)
testBatches = training_set.map(prepareImg).cache().batch(batchSize).map(prepareImg).prefetch(1)
````
